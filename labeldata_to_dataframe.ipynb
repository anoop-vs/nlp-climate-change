```python
with open('/content/project-3-at-2023-03-02-10-36-bf2f0feb (1).json','r') as f:
  labelled_data = json.load(f)
```

```python
#Extract the labled data
labels= []
for d in labelled_data:
  for ann in d['annotations']:
    for result in ann['result']:
        label = result['value']['choices'][0]
        labels.append(label)
        
```

```python
labels
```

```python
texts= []
for d in labelled_data:
        text=d['data']["renderedContent"]
        texts.append(text)
```

```python
import pandas as pd
df=pd.DataFrame(list(zip(texts,labels)),columns=["content","label"])
```
